```python
import datetime
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
import requests
import codecs
import os

wdir = "C:/Users/sc84740/Downloads/W251/"
```

```python
def read_torm_html(year,month,day, counter=0):
    #Sets directory to location of HTML files
    os.chdir(wdir+"tormarket/"+str(year)+"-"+str(month)+"-"+str(day)+"/products")
    #Counter
    c=counter

    #Loops through all files
    for i in os.listdir(os.getcwd()):
        c+=1
        try:
            if i.endswith(".html"):
                #ID = "ID" + str(c+611)
                d=datetime.datetime(year,month,day)
                #Reads in the HTML file
                ecj_data = open(i,'r').read()
                soup = BeautifulSoup(ecj_data)

                #Finds the title of the product
                title = soup.find("title")
                #Finds a table with attributes around the product
                table = soup.find("table",attrs={"class":"table table-striped"})


                #Creates an empty numpy array to hold the data
                #Need to update these parameters for different markets
                data = np.empty((6,2),dtype="S20")

                #Loops through all rows in the table
                count1 =0
                for row in table.find_all("tr"):
                    count2 = 0   
                    #Loops through all columns in the row
                    for td in row.find_all("td"):
                        #Sets the value of the table to the corresponding location in the numpy array.
                        data[count1,count2] = td.get_text()
                        count2+=1
                    count1+=1

                #Converts numeric strings to floats
                data[1,1] = float(re.sub("[^0-9]", "", data.item(3)))
                data[5,1] = float(re.sub("[^0-9]", "", data.item(11)))


                #Creates a pandas data frame with the product title and attributes
                #Title of product
                temp = pd.DataFrame([title.get_text()],index=["Title"],columns=["Values"])
                #Attributes
                temp=temp.append(pd.DataFrame(data[0:,1:], index=data[0:,0],columns=["Values"]))
                #ID and Date variable
                temp=temp.append(pd.DataFrame([d.strftime('%b %d, %Y')],index=["Date"],columns=["Values"]))

                #Transposes the data frame and re-indexes on the custom ID
                temp2=temp.transpose()
                #temp3=temp2.set_index("ID")

                if c==1: output = temp2
                else: output= output.append(temp2)

                if c%100==0: print c
            else:
                continue

        except:
            print "File " + i + " has failed"
            
    return output
```

```python
tor_12_13 = read_torm_html(2013,12,13)
tor_12_14 = read_torm_html(2013,12,14)
```
