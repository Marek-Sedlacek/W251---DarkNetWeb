```python
# This tells matplotlib not to try opening a new window for each plot.
%matplotlib inline

# General libraries.
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# SK-learn libraries for learning.
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.grid_search import GridSearchCV

# SK-learn libraries for evaluation.
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import classification_report

# SK-learn library for importing the newsgroup data.
from sklearn.datasets import fetch_20newsgroups

# SK-learn libraries for feature extraction from text.
from sklearn.feature_extraction.text import *
```

```python
new = pd.read_csv("cleaned_output.csv")

new.head()
```

```python
new2 = new[pd.isnull(new["Cat2"])==False]
categories = ["Stimulants", "Ecstasy", "Cannabis", "Psychadelics", "Relaxants", "Opioids","Sildenafil Citrate","Dissociatives","Analgesics","Steroids/PEDs"]
# new2 = new[new["Cat2"].isin(categories)]
len(new2)
```

```python
new2["Cat2"] = [x if x in categories else "Other" for x in new2["Cat2"]]
#new2
```

```python
data = new2["title"].as_matrix()
labels = new2["Cat2"].as_matrix()

num_test = len(data)
test_data, test_labels = data[:num_test/4], labels[:num_test/4]
train_data, train_labels = data[num_test/4:], labels[num_test/4:]

print num_test
print len(test_data)
print len(train_data)
```

```python
train_labels
```

```python
print "a)"
vectorizer = CountVectorizer()
x=vectorizer.fit_transform(train_data)
print "Size of Vocabulary: " + str(x.shape[1])
print "Average non-zero entries per row: "+ '%.2f' %(float(x.nnz)/x.shape[0])
print "Fraction of entries that are non-zero: " + '%.5f' %(float(x.nnz)/(x.shape[0]*x.shape[1]))
print ""

##e
print "e)"
vectorizer = CountVectorizer(min_df=10)
y=vectorizer.fit_transform(train_data)
print "Size of Vocabulary: " + str(y.shape[1])
print "Average non-zero entries per row: "+ '%.2f' %(float(y.nnz)/y.shape[0])
print ""
```

```python
#Initialize Count Vectorizer
vectorizer = CountVectorizer()
x=vectorizer.fit_transform(train_data)

for i in [0.000001,0.0001,0.005,0.0075,0.01,0.03,0.05,0.1,0.5,1]:
    Mult = MultinomialNB(alpha=i)
    Mult.fit(vectorizer.transform(train_data), train_labels) 
    preds = Mult.predict(vectorizer.transform(test_data))
    
    print "For alpha %f, the f1 score is %f" %(i, metrics.f1_score(test_labels,preds,average="weighted"))
```
